```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/water_potability.csv")
```

```python
df.info()
```

```python
df = df.drop('Potability', axis = 1)
```

```python
df.describe()
```

```python
df = df.fillna(df.mean())
```

```python
df_input = df[['ph', 'Hardness', 'Solids', 'Conductivity', 'Turbidity']]
```

```python
from sklearn.preprocessing import StandardScaler
columns = list(df_input.columns)
scaler = StandardScaler()
df_input[columns] = scaler.fit_transform(df_input)
```

```python
df_input
```

```python
import pickle 
pickle.dump(scaler,open('scaler.pkl', 'wb'))
```

```python
df
```

```python
# chloramines, sulphate, organic_carbon, Trihalomethanes
df1 = pd.concat([df_input, df['Chloramines']], axis=1)
df2 = pd.concat([df_input, df['Sulfate']], axis=1)
df3 = pd.concat([df_input, df['Organic_carbon']], axis=1)
df4 = pd.concat([df_input, df['Trihalomethanes']], axis=1)
```

```python
df1.to_csv("data/chloramines.csv", index= False)
df2.to_csv("data/sulphates.csv", index= False)
df3.to_csv("data/organic_carbon.csv", index= False)
df4.to_csv("data/trihalomethanes.csv", index= False)
```
